```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
openai_api_key = os.environ.get("OPENAI_API_KEY")
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
docs = PyPDFLoader("data/GTB_gold_Nov23.pdf").load()
```

```python
len(docs), docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)

slices = splitter.split_documents(docs)
```

```python
len(slices), slices
```

```python
from langchain_openai.embeddings import OpenAIEmbeddings
```

```python
embedding_model = OpenAIEmbeddings()

embedding_model.model
```

```python
embedding_model.embed_query(slices[0].page_content)
```

```python
from langchain_community.vectorstores import InMemoryVectorStore
```

```python
vectorstore = InMemoryVectorStore.from_documents(
    documents=slices,
    embedding=embedding_model
)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k":2})
```

```python
retriever.invoke("Seguro Viagem")
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "Responda utilizando exclusivamente o conteúdo fornecido: \n\nContexto: \n{contexto}"),
    ("human", "{query}")
])
```

```python
prompt
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
```

```python
model = ChatOpenAI(
    model="gpt-4.1-nano",
    temperature=0.2,
    api_key=openai_api_key
)
```

```python
chain = (
    prompt
    | model
    | StrOutputParser()
)
```

```python
query = "Como lidar com cartão roubado?"
```

```python
context = retriever.invoke(query)
context = "\n\n".join(ctx.page_content for ctx in context)
```

```python
context
```

```python
response = chain.invoke({
    "query": query,
    "contexto": context
})
```
